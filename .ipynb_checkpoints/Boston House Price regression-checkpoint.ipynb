```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python

# Method 1: Direct download from StatLib archive
data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)
data = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
target = raw_df.values[1::2, 2]

# Convert to DataFrame for easier handling
feature_names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 
                 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT']
boston_df = pd.DataFrame(data, columns=feature_names)
boston_df['MEDV'] = target

print(f"Boston housing dataset has {boston_df.shape[0]} data points with {boston_df.shape[1]} variables each.")
```

```python
def plot_correlation_heatmap(df, figsize=(10, 8), cmap='coolwarm', annot=True, fmt='.2f'):
    """
    Create and plot a correlation matrix heatmap for a given DataFrame.
    
    Parameters:
    df (pandas.DataFrame): Input DataFrame
    figsize (tuple): Figure size (width, height) in inches
    cmap (str): Color map for the heatmap
    annot (bool): If True, write the data value in each cell
    fmt (str): String formatting code to use when adding annotations
    """
    # Create correlation matrix
    correlation_matrix = df.corr()
    
    # Create the heatmap
    plt.figure(figsize=figsize)
    sns.heatmap(correlation_matrix, annot=annot, cmap=cmap, fmt=fmt)
    
    plt.title('Correlation Matrix Heatmap')
    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def analyze_data_distribution(df, figsize=(20, 15)):
    """
    Analyze and visualize the data distribution for each column in a DataFrame.
    
    Parameters:
    df (pandas.DataFrame): Input DataFrame
    figsize (tuple): Figure size (width, height) in inches
    
    Returns:
    pandas.DataFrame: Summary statistics for each column
    """
    # Calculate summary statistics
    summary = df.describe(percentiles=[.25, .5, .75])
    summary.loc['std'] = df.std()  # Add standard deviation
    summary = summary.T  # Transpose for better readability
    
    # Prepare for plotting
    num_cols = df.select_dtypes(include=[np.number]).columns
    num_features = len(num_cols)
    
    # Handle case with a single feature
    if num_features == 1:
        fig, axes = plt.subplots(1, 2, figsize=figsize)
        col = num_cols[0]
        
        # Box plot
        sns.boxplot(x=df[col], ax=axes[0])
        axes[0].set_title(f'Box Plot: {col}', fontsize=14)
        axes[0].set_xlabel('')
        axes[0].tick_params(axis='both', labelsize=12)
        
        # Violin plot
        sns.violinplot(x=df[col], ax=axes[1])
        axes[1].set_title(f'Violin Plot: {col}', fontsize=14)
        axes[1].set_xlabel('')
        axes[1].tick_params(axis='both', labelsize=12)
    else:
        # Create subplots with adjusted layout
        fig, axes = plt.subplots(num_features, 2, figsize=figsize)
        
        for i, col in enumerate(num_cols):
            # Box plot
            sns.boxplot(x=df[col], ax=axes[i, 0])
            axes[i, 0].set_title(f'Box Plot: {col}', fontsize=14)
            axes[i, 0].set_xlabel('')
            axes[i, 0].tick_params(axis='both', labelsize=12)
            
            # Violin plot
            sns.violinplot(x=df[col], ax=axes[i, 1])
            axes[i, 1].set_title(f'Violin Plot: {col}', fontsize=14)
            axes[i, 1].set_xlabel('')
            axes[i, 1].tick_params(axis='both', labelsize=12)
    
    fig.suptitle('Data Distribution Analysis', fontsize=20, y=0.98)
    plt.subplots_adjust(hspace=0.4, wspace=0.3)  # Add more space between plots
    plt.tight_layout(rect=[0, 0, 1, 0.97])  # Adjust layout, leaving room for suptitle
    plt.show()
    
    # Print summary statistics in a more readable format
    print("\n=== SUMMARY STATISTICS ===")
    print(summary)
    
    return summary
```

```python
# Call the function with the sample DataFrame
summary_stats = analyze_data_distribution(boston_df)
print(summary_stats)
```

```python
boston_df = scaler.fit_transform(boston_df)
```

```python
X = boston_df[:, :-1]
y = boston_df[:, -1]
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import KFold, train_test_split
from sklearn.metrics import mean_squared_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

def build_and_evaluate_regression_model(X, y, n_folds=5, epochs=100, batch_size=5, validation_split=0.2):
    """
    Build, train and evaluate a neural network for regression using k-fold cross-validation.
    
    Parameters:
    X (array-like): Input features
    y (array-like): Target values
    n_folds (int): Number of folds for cross-validation
    epochs (int): Number of training epochs
    batch_size (int): Batch size for training
    validation_split (float): Fraction of training data to use for validation
    
    Returns:
    tuple: (average_train_mse, average_train_rmse, average_val_mse, average_val_rmse, predictions, actual_values)
    """
    # Initialize lists to store metrics
    train_mse_scores, train_rmse_scores = [], []
    val_mse_scores, val_rmse_scores = [], []
    all_predictions, all_actual = [], []
    all_train_loss, all_val_loss = [], []
    
    # Define the K-fold Cross Validator
    kfold = KFold(n_splits=n_folds, shuffle=True, random_state=42)
    
    # K-fold Cross Validation
    fold_no = 1
    for train_val, test in kfold.split(X, y):
        # Split train_val into train and validation sets
        X_train, X_val, y_train, y_val = train_test_split(X[train_val], y[train_val], test_size=validation_split, random_state=42)
        
        # Build the model
        model = Sequential([
            Dense(20, activation='relu', input_dim=X.shape[1]),
            # Dense(6, activation='relu', input_dim=X.shape[1]),
            Dense(1, activation='linear')
        ])
        
        # Compile the model
        model.compile(optimizer='adam', loss='mse', metrics=['mse'])
        
        # Print model architecture for the first fold
        if fold_no == 1:
            model.summary()
        
        # Train the model
        history = model.fit(
            X_train, y_train,
            validation_data=(X_val, y_val),
            epochs=epochs,
            batch_size=batch_size,
            verbose=0
        )
        
        # Store training history
        all_train_loss.extend(history.history['loss'])
        all_val_loss.extend(history.history['val_loss'])
        
        # Evaluate the model
        train_predictions = model.predict(X_train, verbose=0)
        val_predictions = model.predict(X_val, verbose=0)
        test_predictions = model.predict(X[test], verbose=0)
        
        train_mse = mean_squared_error(y_train, train_predictions)
        train_rmse = np.sqrt(train_mse)
        val_mse = mean_squared_error(y_val, val_predictions)
        val_rmse = np.sqrt(val_mse)
        
        # Store metrics
        train_mse_scores.append(train_mse)
        train_rmse_scores.append(train_rmse)
        val_mse_scores.append(val_mse)
        val_rmse_scores.append(val_rmse)
        all_predictions.extend(test_predictions.flatten())
        all_actual.extend(y[test])
        
        print(f'Fold {fold_no}:')
        print(f'  Train MSE = {train_mse:.4f}, Train RMSE = {train_rmse:.4f}')
        print(f'  Val MSE = {val_mse:.4f}, Val RMSE = {val_rmse:.4f}')
        fold_no += 1
    
    # Calculate average metrics
    avg_train_mse = np.mean(train_mse_scores)
    avg_train_rmse = np.mean(train_rmse_scores)
    avg_val_mse = np.mean(val_mse_scores)
    avg_val_rmse = np.mean(val_rmse_scores)
    
    print(f'\nAverage Train MSE: {avg_train_mse:.4f}, Average Train RMSE: {avg_train_rmse:.4f}')
    print(f'Average Val MSE: {avg_val_mse:.4f}, Average Val RMSE: {avg_val_rmse:.4f}')
    
    # Plot training and validation loss
    plt.figure(figsize=(10, 4))
    plt.subplot(1, 2, 1)
    plt.plot(all_train_loss, label='Training Loss')
    plt.plot(all_val_loss, label='Validation Loss')
    plt.title('Training and Validation Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    
    # Plot predicted vs actual values
    plt.subplot(1, 2, 2)
    plt.scatter(all_actual, all_predictions, alpha=0.5)
    plt.plot([min(all_actual), max(all_actual)], [min(all_actual), max(all_actual)], 'r--')
    plt.title('Predicted vs Actual')
    plt.xlabel('Actual Values')
    plt.ylabel('Predicted Values')
    
    plt.tight_layout()
    plt.show()
    
    return avg_train_mse, avg_train_rmse, avg_val_mse, avg_val_rmse, all_predictions, all_actual

# Example usage:
# avg_train_mse, avg_train_rmse, avg_val_mse, avg_val_rmse, predictions, actual = build_and_evaluate_regression_model(X, y)
```
