```python
# Create your first MLP in Keras
from keras.models import Sequential
from keras.layers import Dense
import numpy as np
import pandas as pd
import requests
import os

# Fix random seed for reproducibility
seed = 7
np.random.seed(seed)

# Download the Pima Indians dataset
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv"
filename = "pima-indians-diabetes.csv"
if not os.path.exists(filename):
    response = requests.get(url)
    with open(filename, 'wb') as file:
        file.write(response.content)
    print(f"Dataset downloaded to {filename}")
else:
    print(f"Dataset already exists at {filename}")

# Load Pima Indians Diabetes dataset
dataset = np.loadtxt(filename, delimiter=",")
# Split into input (X) and output (Y) variables
X = dataset[:, 0:8]
Y = dataset[:, 8]

# Create the model
model = Sequential()
model.add(Dense(12, input_dim=8, kernel_initializer='uniform', activation='relu'))
model.add(Dense(8, kernel_initializer='uniform', activation='relu'))
model.add(Dense(1, kernel_initializer='uniform', activation='sigmoid'))

# Compile the model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Fit the model
model.fit(X, Y, epochs=150, batch_size=10)

# Evaluate the model
scores = model.evaluate(X, Y)
print("%s: %.2f%%" % (model.metrics_names[1], scores[1] * 100))
```

```python
# MLP for Pima Indians Dataset with Grid Search via sklearn
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import GridSearchCV
import numpy as np
import requests
import os

# Function to create model, required for KerasClassifier
def create_model(optimizer='rmsprop', kernel_initializer='glorot_uniform'):
    # Create model
    model = Sequential()
    model.add(Dense(12, input_dim=8, kernel_initializer=kernel_initializer, activation='relu'))
    model.add(Dense(8, kernel_initializer=kernel_initializer, activation='relu'))
    model.add(Dense(1, kernel_initializer=kernel_initializer, activation='sigmoid'))
    # Compile model
    model.compile(loss='binary_crossentropy', optimizer=optimizer, metrics=['accuracy'])
    return model

# Fix random seed for reproducibility
seed = 7
np.random.seed(seed)

# Download the Pima Indians dataset
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv"
filename = "pima-indians-diabetes.csv"
if not os.path.exists(filename):
    response = requests.get(url)
    with open(filename, 'wb') as file:
        file.write(response.content)
    print(f"Dataset downloaded to {filename}")
else:
    print(f"Dataset already exists at {filename}")

# Load Pima Indians Diabetes dataset
dataset = np.loadtxt(filename, delimiter=",")
# Split into input (X) and output (Y) variables
X = dataset[:, 0:8]
Y = dataset[:, 8]

# Create model for KerasClassifier
model = KerasClassifier(build_fn=create_model, verbose=0)

# Grid search hyperparameters
optimizers = ['rmsprop', 'adam']  # Optimizers
kernel_initializers = ['glorot_uniform', 'normal', 'uniform']  # Initializers
epochs = [50, 100, 150]  # Number of epochs
batches = [5, 10, 20]  # Batch sizes

# Create the grid search parameter dictionary
param_grid = dict(optimizer=optimizers, kernel_initializer=kernel_initializers, epochs=epochs, batch_size=batches)

# Perform the grid search
grid = GridSearchCV(estimator=model, param_grid=param_grid, cv=3)  # 3-fold cross-validation
grid_result = grid.fit(X, Y)

# Summarize results
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
for mean, std, params in zip(grid_result.cv_results_['mean_test_score'], grid_result.cv_results_['std_test_score'], grid_result.cv_results_['params']):
    print("%f (%f) with: %r" % (mean, std, params))
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV, KFold
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier
from tensorflow.keras.utils import to_categorical
import requests
import os

# Step 1: Download the Iris dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
filename = "iris.csv"

# Download the dataset if it doesn't exist
if not os.path.exists(filename):
    response = requests.get(url)
    with open(filename, "wb") as file:
        file.write(response.content)
    print(f"Dataset downloaded to {filename}")
else:
    print(f"Dataset already exists at {filename}")

# Step 2: Load the dataset
column_names = ["sepal_length", "sepal_width", "petal_length", "petal_width", "class"]
data = pd.read_csv(filename, header=None, names=column_names)

# Print the head of the dataset
print("Head of the dataset:")
print(data.head())

# Step 3: Normalize the features
features = data.iloc[:, :-1].values
scaler = MinMaxScaler()
normalized_features = scaler.fit_transform(features)

# Step 4: Encode the target column (classes)
target = data["class"].values
label_encoder = LabelEncoder()
integer_encoded = label_encoder.fit_transform(target)  # Assign 0, 1, 2 for the classes
one_hot_encoded_target = to_categorical(integer_encoded)  # Create one-hot encoding for classes

# Step 5: Split into train and test sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(normalized_features, integer_encoded, test_size=0.2, random_state=42)

# Step 6: Define a function to create the neural network model
def create_model(optimizer='adam'):
    model = Sequential()
    model.add(Dense(36, input_dim=X_train.shape[1], activation='relu'))  # Input layer with 36 neurons
    model.add(Dense(12, activation='relu'))  # Hidden layer with 12 neurons
    model.add(Dense(3, activation='softmax'))  # Output layer with 3 neurons (one for each class)
    model.compile(loss='sparse_categorical_crossentropy', optimizer=optimizer, metrics=['accuracy'])
    return model

# Step 7: K-Fold Cross-Validation with Grid Search
model = KerasClassifier(build_fn=create_model, verbose=0)

# Define the grid of hyperparameters to search
param_grid = {
    'batch_size': [10, 20],
    'epochs': [50, 100],
    'optimizer': ['adam', 'rmsprop']
}

# Set up K-Fold cross-validation
kfold = KFold(n_splits=5, shuffle=True, random_state=42)
grid = GridSearchCV(estimator=model, param_grid=param_grid, cv=kfold, n_jobs=-1)

# Perform the grid search on the training data
grid_result = grid.fit(X_train, y_train)

# Step 8: Summarize the grid search results
print("Best Parameters: ", grid_result.best_params_)
print("Best Cross-Validation Accuracy: %.2f%%" % (grid_result.best_score_ * 100))
for mean, std, params in zip(grid_result.cv_results_['mean_test_score'], grid_result.cv_results_['std_test_score'], grid_result.cv_results_['params']):
    print("Accuracy: %.2f%% (+/- %.2f%%) with: %r" % (mean * 100, std * 100, params))

# Step 9: Evaluate the best model on the test set
best_model = grid_result.best_estimator_
test_loss, test_accuracy = best_model.model.evaluate(X_test, y_test, verbose=0)
print(f"Test Loss: {test_loss:.4f}, Test Accuracy: {test_accuracy:.4f}")
```
